```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the CSV file
df = pd.read_csv('/content/Student_Cleaned.csv')

# Display the first few rows of the DataFrame to understand its structure
print('First 5 rows of the dataset:')
display(df.head())

# Display basic information about the DataFrame, including column names and data types
print('\nDataFrame Information:')
df.info()
```

### Scatter Plot: Hours Studied vs. Performance Index

This plot shows the relationship between the number of 'Hours Studied' and the 'Performance Index' for each student. Each dot represents a student.

```python
# Create a scatter plot
plt.figure(figsize=(10, 6)) # Set the size of the plot
plt.scatter(df['Hours Studied'], df['Performance Index'], alpha=0.6) # Plot points with some transparency

plt.title('Student Performance Index vs. Hours Studied') # Title of the plot
plt.xlabel('Hours Studied') # Label for the x-axis
plt.ylabel('Performance Index') # Label for the y-axis
plt.grid(True) # Add a grid for better readability
plt.show() # Display the plot
```

### Bar Chart: Distribution of Extracurricular Activities

This bar chart shows how many students participate in 'Extracurricular Activities' versus how many do not.

```python
# Count how many students fall into each category of 'Extracurricular Activities'
activity_counts = df['Extracurricular Activities'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 5)) # Set the size of the plot
plt.bar(activity_counts.index, activity_counts.values, color='skyblue') # Plot bars

plt.title('Number of Students by Extracurricular Activities') # Title of the plot
plt.xlabel('Extracurricular Activities') # Label for the x-axis
plt.ylabel('Number of Students') # Label for the y-axis
plt.grid(axis='y') # Add a horizontal grid for better readability
plt.show() # Display the plot
```

### Line Chart: Average Performance Index by Hours Studied

This line chart shows how the average 'Performance Index' changes as the 'Hours Studied' increases.

```python
# Calculate the average 'Performance Index' for each unique 'Hours Studied' value
avg_performance_by_study_hours = df.groupby('Hours Studied')['Performance Index'].mean().reset_index()

# Create a line chart
plt.figure(figsize=(10, 6)) # Set the size of the plot
plt.plot(
    avg_performance_by_study_hours['Hours Studied'],
    avg_performance_by_study_hours['Performance Index'],
    marker='o', # Show data points with circles
    linestyle='-', # Connect points with a solid line
    color='red' # Set the line color to red
)

plt.title('Average Performance Index Trend by Hours Studied') # Title of the plot
plt.xlabel('Hours Studied') # Label for the x-axis
plt.ylabel('Average Performance Index') # Label for the y-axis
plt.grid(True) # Add a grid for better readability
plt.show() # Display the plot
```
